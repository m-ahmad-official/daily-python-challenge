```python
from PIL import Image
import numpy as np

# ASCII characters ordered by brightness (dark to light)
ASCII_CHARS = "@#*+-. "

def image_to_ascii(image_path, width=100):
    """Convert an image to ASCII art."""
    
    # Open and convert image to grayscale
    img = Image.open(image_path).convert("L") 
    
    # Resize image to maintain aspect ratio
    aspect_ratio = img.height / img.width
    new_height = int(width * aspect_ratio * 0.5)  # Adjust for character aspect ratio
    img = img.resize((width, new_height))
    
    # Convert pixels to ASCII characters
    pixels = np.array(img)
    ascii_image = "\n".join("".join(ASCII_CHARS[int(pixel / 255 * (len(ASCII_CHARS) - 1))] for pixel in row) for row in pixels)
    
    return ascii_image

# Load image and convert to ASCII
image_path = input("Enter image path: ")
ascii_art = image_to_ascii(image_path)

# Output ASCII Art
print("\nðŸ”³ ASCII Art:\n")
print(ascii_art)

# Save to a text file
with open("ascii_art.txt", "w") as file:
    file.write(ascii_art)

print("\nâœ… ASCII Art saved to 'ascii_art.txt'")
```
