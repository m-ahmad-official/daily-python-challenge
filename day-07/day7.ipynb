```python
def main():
    tasks = []  # Task list
    
    while True:
        print("\nğŸ“Œ Welcome to To-Do List!")
        print("1. Add Task")
        print("2. Show Tasks ")
        print("3. Delete Task")
        print("4. Exit")
        
        choice = input("Choose an option (1-4): ")
        print(f"Choose an option (1-4): {choice}")
        
        if choice == "1":
            task = input("Enter new task: ")
            print(f"Enter new task: {task}")
            tasks.append(task)
            print("Task Added! âœ…")
        
        elif choice == "2":
            if tasks:
                print("\nğŸ“ Your To-Do List:")
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")
            else:
                print("No tasks found! ğŸ“")
        
        elif choice == "3":
            if tasks:
                print("\nğŸ“ Your To-Do List:")
                for index, task in enumerate(tasks, start=1):
                    print(f"{index}. {task}")
                try:
                    task_index = int(input("Enter task number to delete: ")) - 1
                    print(f"Enter task number to delete: {task_index + 1}")
                    if 0 <= task_index < len(tasks):
                        removed_task = tasks.pop(task_index)
                        print(f"Task '{removed_task}' deleted! âŒ")
                    else:
                        print("Invalid task number! ğŸš«")
                except ValueError:
                    print("Please enter a valid number! ğŸ”¢")
            else:
                print("No tasks to delete! ğŸ“")
        
        elif choice == "4":
            print("Exiting... Goodbye! ğŸ‘‹")
            break
        
        else:
            print("Invalid choice! Please choose between 1-4. ğŸš€")

if __name__ == "__main__":
    main()
```
